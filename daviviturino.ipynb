```python
numero = float(input("Digite um número: "))

if numero > 10:
  print(f"O número {numero} é maior que 10.")
elif numero == 10:
  print(f"O número {numero} é igual a 10.")
else:
  print(f"O número {numero} é menor ou igual a 10.")
```

```python
numero = int(input("Digite um número inteiro: "))

if numero % 2 == 0:
  print(f"O número {numero} é par.")
else:
  print(f"O número {numero} é ímpar.")
```

```python
nome = input("Digite um nome: ")

if "a" in nome or "A" in nome:
  print(f"O nome '{nome}' contém a letra 'a'.")
else:
  print(f"O nome '{nome}' não contém a letra 'a'.")
```

```python
numero = int(input("Digite um número inteiro: "))

if numero % 5 == 0:
  print(f"O número {numero} é múltiplo de 5.")
else:
  print(f"O número {numero} não é múltiplo de 5.")
```

```python
temperatura = float(input("Digite a temperatura atual em graus Celsius: "))

if temperatura > 30:
  print(f"A temperatura atual ({temperatura}°C) está quente.")
else:
  print(f"A temperatura atual ({temperatura}°C) está frio.")
```

```python
entrada = input("Digite 'S' para continuar ou 'N' para sair: ")

if entrada.upper() == "S":
  print("Você escolheu continuar.")
elif entrada.upper() == "N":
  print("Você escolheu sair.")
else:
  print("Entrada inválida. Por favor, digite 'S' ou 'N'.")
```

```python
numero = float(input("Digite um número: "))

if numero < 0:
  print(f"O número {numero} é negativo.")
elif numero == 0:
  print(f"O número {numero} é zero.")
else:
  print(f"O número {numero} é positivo.")
```

```python
numero_dia = int(input("Digite um número de 1 a 7: "))

if numero_dia == 1:
  print("O dia correspondente é Domingo.")
elif numero_dia == 2:
  print("O dia correspondente é Segunda-feira.")
elif numero_dia == 3:
  print("O dia correspondente é Terça-feira.")
elif numero_dia == 4:
  print("O dia correspondente é Quarta-feira.")
elif numero_dia == 5:
  print("O dia correspondente é Quinta-feira.")
elif numero_dia == 6:
  print("O dia correspondente é Sexta-feira.")
elif numero_dia == 7:
  print("O dia correspondente é Sábado.")
else:
  print("Número inválido. Por favor, digite um número de 1 a 7.")
```

```python
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
  print(f"O ano {ano} é bissexto.")
else:
  print(f"O ano {ano} não é bissexto.")
```

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
  salario = horas_trabalhadas * valor_por_hora
else:
  horas_normais = 40
  horas_extras = horas_trabalhadas - horas_normais
  salario = (horas_normais * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

print(f"O salário é: R${salario:.2f}")
```
